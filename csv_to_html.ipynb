```python
# Import Depdendencies
import pandas as pd
from datetime import datetime
from html import escape
```

```python
# Read the csv file into a dataframe
cities_df = pd.read_csv('./assets/data/cities.csv')
cities_df.head()
```

```python
# Sort the DataFrame by City Name into a new DataFrame
cities_sorted_df = cities_df.sort_values(by='City').copy()
```

```python
# Reset the index of the sorted DataFrame
cities_sorted_df.reset_index(level=0, inplace=True)
cities_sorted_df.head()
```

```python
# Create a new DataFrame with index starting at 1 and Date converted to string format ('YYYY-MM-DD')
cities_cleaned_df = pd.DataFrame(
    [ index+1
     ,row['City'].title()
     ,row['Country']
     #,datetime.utcfromtimestamp(row['Date']).strftime('%Y-%m-%d')
     ,row['Date']
     ,row['Lat']
     ,row['Lng']
     ,row['Max Temp']
     ,row['Humidity']
     ,row['Cloudiness']
     ,row['Wind Speed']
    ] for index, row in cities_sorted_df.iterrows()
)
cities_cleaned_df.columns = [ '#'
                             ,'City'
                             ,'Country'
                             ,'Date'
                             ,'Latitude'
                             ,'Longitude'
                             ,'Max Temp'
                             ,'Humidity'
                             ,'Cloudiness'
                             ,'Wind Speed'
]

cities_cleaned_df.head()
```

```python
cities_html = cities_cleaned_df.to_html()
cities_html
```

# Steps to process the data in preparation for Bootstrap based table object for formatting and display

```python
# Get the column Names
column_names = []
for index, row in cities_cleaned_df.head().iterrows():
    for name in row.keys():
        if name not in column_names:
            column_names.append(name)
```

```python
# Create a new dictionary 
# set each column to be either 'L'eft or 'R'ight aligned
newlist = {}
for x in range(1):
    newlist[x]={}
    for y in column_names:
        newlist[x][y] = 'L'
        
# Get a list frim the Dictionary, bring only the 0'th record
newlist[0]
my_list = [(k,v) for k,v in newlist[0].items()]
```

```python
# Conver the list to a DataFrame for further processing
df = pd.DataFrame(my_list)
df.columns = ['ColumnName', 'Align']

# Manually analysze and identify the columns that will be right or left aligned for display
df.loc[df['ColumnName'] == 'Latitude', ['Align']] = ['R']
df.loc[df['ColumnName'] == 'Longitude', ['Align']] = ['R']
df.loc[df['ColumnName'] == 'Max Temp', ['Align']] = ['R']
df.loc[df['ColumnName'] == 'Humidity', ['Align']] = ['R']
df.loc[df['ColumnName'] == 'Cloudiness', ['Align']] = ['R']
df.loc[df['ColumnName'] == 'Wind Speed', ['Align']] = ['R']

df = df.set_index("ColumnName", drop=False)
df.head()
```

```python
column_names
```

```python
# Now that we have the cleaned table data, the conversion of thedata to HTML can proceed.
#Process data and create the HTML table header row
lines = []
lines.append('<table class="table table-hover table-sm table-striped table-bordered">\n')
lines.append('  <thead class="thead-dark">\n')
lines.append('    <tr>\n')

for name in df.ColumnName:
    if df.loc[df['ColumnName'] == name]['Align'][0] == 'L':
        lines.append(f'      <th class="text-left" scope="col">{escape(name)}</th>\n')
    else:
        lines.append(f'      <th class="text-right" scope="col">{escape(name)}</th>\n')
lines.append('    </tr>\n')
lines.append('  <thead>\n')

#Process data and create the HTML data row
for index, row in cities_cleaned_df.iterrows():
    lines.append('  <tr>\n')
    #for name in column_names:
    for name in df.ColumnName:
        value = row.get(name, '')
        #If the column id the ID column then generate a 'th' row for the 'row' left aligned
        if name == '#':
            lines.append(f'      <th class="text-left" scope="row">{escape(str(value))}</th>\n')
        else:
            #for all other columns determine the alignment and generate the td row with proper alignment
            if df.loc[df['ColumnName'] == name]['Align'][0] == 'L':
                lines.append(f'      <td class="text-left" scope="row">{escape(str(value))}</td>\n')
            else:
                lines.append(f'      <td class="text-right" scope="row">{escape(str(value))}</td>\n')
    lines.append('  </tr>\n')
# Append all the rows generated    
lines.append('</table>')

# The below step merges all the lines into one single line
''.join(lines)    
```

```python
# Write the html file to the file system
# this is an HTML file and will need to be incorated in a displayable html page using VS Code
with open('./assets/data/cities.html', 'w') as f:
    for line in lines:
        f.write("%s" % line)
```
